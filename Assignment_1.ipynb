## 3.	function: 試寫一函數於以下空格。此function名稱為Arithmetic (m,n)，參數為m,n，請將   m,n 傳入function中，如果m>n，回傳m*n，如果m<n，回傳m/n，如果m=n，回傳1。(20%)

```python
def Arithmetic(m,n) :
    if m > n :
        return m * n
    if m < n :
        return m / n 
    if m == n :
        return 1
```
